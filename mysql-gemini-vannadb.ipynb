## Training
You only need to train once. Do not train again unless you want to add more training data.

```python

# The information schema query may need some tweaking depending on your database. This is a good starting point.
df_information_schema = vn.run_sql("SELECT * FROM INFORMATION_SCHEMA.COLUMNS")

# This will break up the information schema into bite-sized chunks that can be referenced by the LLM
plan = vn.get_training_plan_generic(df_information_schema)
plan

# If you like the plan, then uncomment this and run it to train
# vn.train(plan=plan)

```

```python

# The following are methods for adding training data. Make sure you modify the examples to match your database.

# DDL statements are powerful because they specify table names, colume names, types, and potentially relationships
vn.train(ddl="""
CREATE TABLE `aircrafts_data` (
  `aircraft_code` char(3) NOT NULL COMMENT 'Code for the aircraft',
  `model` varchar(150) DEFAULT NULL COMMENT 'Aircraft model in JSON format',
  `range` int DEFAULT NULL COMMENT 'The range of the aircraft',
  PRIMARY KEY (`aircraft_code`)
)
""")
# Sometimes you may want to add documentation about your business terminology or definitions.
#vn.train(documentation="aircrafts_data this table provides information about the aircraft")

vn.train(ddl="""
CREATE TABLE `airports_data` (
  `airport_code` char(3) NOT NULL COMMENT 'Code for the airport',
  `airport_name` varchar(150) DEFAULT NULL COMMENT 'Name of the airport in JSON format',
  `city` varchar(150) DEFAULT NULL COMMENT 'City where the airport is located',
  `coordinates` varchar(255) DEFAULT NULL COMMENT 'Geographic coordinates of the airport',
  `timezone` varchar(255) DEFAULT NULL COMMENT 'Timezone of the airport',
  PRIMARY KEY (`airport_code`)
)
""")
#vn.train(documentation="aircrafts_data this table provides information about the aircraft")

vn.train(ddl="""
CREATE TABLE `flights` (
  `flight_id` int NOT NULL COMMENT 'Flight ID',
  `flight_no` char(6) DEFAULT NULL COMMENT 'Flight number',
  `scheduled_departure` varchar(30) DEFAULT NULL COMMENT 'Scheduled departure time with timestamp and time zone',
  `scheduled_arrival` varchar(30) DEFAULT NULL COMMENT 'Scheduled arrival time with timestamp and time zone',
  `departure_airport` char(3) DEFAULT NULL COMMENT 'Departure airport code',
  `arrival_airport` char(3) DEFAULT NULL COMMENT 'Arrival airport code',
  `status` varchar(20) DEFAULT NULL COMMENT 'Flight status(Enumerated value:Arrived,Scheduled,On Time,Delayed,Cancelled,Departed)',
  `aircraft_code` char(3) DEFAULT NULL COMMENT 'Aircraft code',
  `actual_departure` varchar(30) DEFAULT NULL COMMENT 'Actual departure time with timestamp and time zone',
  `actual_arrival` varchar(30) DEFAULT NULL COMMENT 'Actual arrival time with timestamp and time zone',
  PRIMARY KEY (`flight_id`)
)
""")

# Sometimes you may want to add documentation about your business terminology or definitions.
#vn.train(documentation="Our business defines OTIF score as the percentage of orders that are delivered on time and in full")

# You can also add SQL queries to your training data. This is useful if you have some queries already laying around. You can just copy and paste those from your editor to begin generating new SQL.
#vn.train(sql="SELECT * FROM my-table WHERE name = 'John Doe'")

vn.train(question="What is the range of the 319 airport", sql="SELECT range FROM aircrafts_data WHERE airport_code = '319'")
vn.train(question="Show me ABA airport's information", sql="SELECT * FROM airports_data WHERE airport_code = 'ABA'")
vn.train(question="Query all flight information for 319", sql="SELECT * FROM flights where aircraft_code='319'")
vn.train(question="How many flights are delayed from Moscow to Chelyabinsk", sql="SELECT count(*) FROM `flights` where departure_airport in (select airport_code from airports_data where city='Moscow') and arrival_airport in (select airport_code from airports_data where city='Chelyabinsk') and status='Delayed'")

```

```python
# At any time you can inspect what training data the package is able to reference
training_data = vn.get_training_data()
training_data
```
